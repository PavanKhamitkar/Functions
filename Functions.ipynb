1. What is the difference between a function and a method in Python?
  - A function is defined independently using def, while a method is a function that belongs to an object (usually defined inside a class).


```python
def greet(): return "Hello"          # Function
print("hello".upper())                      # Method

print(greet())
```

2. Explain the concept of function arguments and parameters in Python.
  - Parameters are variables listed in the function definition; arguments are the actual values passed to the function during a call.

```python
def add(a, b): return a + b         # a, b = parameters
add(3, 5)                           # 3, 5 = arguments
```

3. What are the different ways to define and call a function in Python?
  - Functions are defined using def or lambda, and called by their name followed by ().

```python
def square(x): return x*x
print("Using def - "+str(square(4)))                           # Using def

print("Using lambda - "+str((lambda x: x*x)(4)))                  # Using lambda
```

4. What is the purpose of the `return` statement in a Python function?
  - `return` sends a result back to the caller and exits the function.

```python
def double(x): return x * 2
print(double(3))                    # Output: 6
```

5. What are iterators in Python and how do they differ from iterables?
  - An iterable can be looped over, while an iterator is an object that returns elements using `next()`.

```python
nums = [1, 2, 3]                    # Iterable
it = iter(nums)                    # Iterator
print(next(it))                    # Output: 1
```

6. Explain the concept of generators in Python and how they are defined.
  - Generators are functions that yield items one at a time using `yield`, saving memory.

```python
def gen():
    yield 1
    yield 2
print(list(gen()))                 # Output: [1, 2]
```

9. Explain the purpose and usage of the `map()` function in Python

```python
print(list(map(lambda x: x*2, [1, 2, 3])))  # Output: [2, 4, 6]
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - `map()` transforms, `filter()` selects, `reduce()` aggregates values.

```python
from functools import reduce
print(list(filter(lambda x: x%2==0, [1,2,3])))     # [2]
print(reduce(lambda x,y: x+y, [1,2,3]))            # 6
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
my_list = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(my_list))  # Output: 12
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2 or n > 200:
        return "Please enter a number between 1 and 200."
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return f"{n} is not a prime number."
    return f"{n} is a prime number."

# Example usage:
print(is_prime(97))   # Output: 97 is a prime number.
print(is_prime(100))  # Output: 100 is not a prime number.
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

# Example usage:
for val in powers_of_two(5):
    print(val, end=" ")
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_str = "Hello, Sumeet!"
filtered_str = ''.join(filter(is_not_vowel, input_str))

print(filtered_str)
```

```python
def calculate_order_values():
    # Sample data from the book shop
    orders = [
        {"order_number": "34587", "title": "Learning Python, Mark Lutz", "quantity": 4, "price": 40.95},
        {"order_number": "98762", "title": "Programming Python, Mark Lutz", "quantity": 5, "price": 56.80},
        {"order_number": "77226", "title": "Head First Python, Paul Barry", "quantity": 3, "price": 32.95},
        {"order_number": "88112", "title": "Einfhrung in Python3, Bernd Klein", "quantity": 3, "price": 24.99}
    ]

    # Using lambda and map to calculate the order values
    # For each order, we calculate the total and add a 10 surcharge if the total is less than 100
    result = list(map(
        lambda order: (
            order["order_number"],
            order["quantity"] * order["price"] + 10 if order["quantity"] * order["price"] < 100 else order["quantity"] * order["price"]
        ),
        orders
    ))

    return result

# Test the function
if __name__ == "__main__":
    order_values = calculate_order_values()
    print("Order Number | Order Value ()")
    print("-" * 30)
    for order_num, value in order_values:
        print(f"{order_num:<12} | {value:.2f}")
```
